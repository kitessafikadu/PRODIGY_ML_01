```python
from sklearn.preprocessing import LabelEncoder, StandardScaler, PolynomialFeatures
from sklearn.model_selection import train_test_split
```

```python
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
```

```python
train.drop(['PoolQC' , 'Fence', 'Alley'], axis=1, inplace=True)
```

```python
test.drop(['PoolQC' , 'Fence', 'Alley'], axis=1, inplace=True)
```

```python
# train.columns
# test.columns
```

```python
train.fillna(train.select_dtypes(include=["number"]).median(), inplace=True)
test.fillna(test.select_dtypes(include=["number"]).median(), inplace=True)

categorical_cols = train.select_dtypes(include=["object"]).columns
mode_values = train[categorical_cols].mode().iloc[0]

train.fillna(mode_values, inplace=True)
test.fillna(mode_values, inplace=True)
```

```python
train.isna().any().sum()
```

```python
train.drop('Id', axis=1, inplace=True)
test.drop('Id', axis=1, inplace=True)
train
```

```python
X = train.select_dtypes(include=["object"]).columns
X
```

```python
label_encoders = {}
for column in X:
    le = LabelEncoder()
    train[column] = le.fit_transform(train[column])
    test[column] = le.fit_transform(test[column])
    label_encoders[column] = le
    
```

```python
train.shape, test.shape
```

```python
x = np.array(train.drop("SalePrice", axis=1))
y = np.array(train["SalePrice"])
x.shape,y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(
 x, y, test_size=0.33, random_state=42)
```

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
poly_reg = PolynomialFeatures(degree=2)
x_poly_train = poly_reg.fit_transform(X_train)
x_poly_test = poly_reg.transform(X_test)
```

```python
train.shape, x.shape
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_poly_train, y_train)
y_pred = model.predict(x_poly_test)
```

```python
from mlxtend.evaluate import bias_variance_decomp
avg_error,avg_bias,avg_variance =  bias_variance_decomp(model,x_poly_train,y_train,x_poly_test,y_test,loss='mse',random_seed=1)
avg_error,avg_bias,avg_variance
```
